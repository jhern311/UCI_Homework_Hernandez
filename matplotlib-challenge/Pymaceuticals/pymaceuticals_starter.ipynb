```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from sklearn import datasets

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_study = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
combined_study
```

```python
# Checking the number of mice.
number_mice = combined_study['Mouse ID'].value_counts()
raw_number_mice = len(number_mice)
raw_number_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_id = combined_study.loc[combined_study.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
dup_mouse_id = pd.DataFrame({"Mouse ID": duplicate_id})
dup_mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_df = combined_study[combined_study['Mouse ID'].isin(duplicate_id)]
duplicate_mouse_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# clean_dataset = combined_study.drop(combined_study.index[[908, 909, 910, 911, 912, 913, 914, 915, 916, 917, 918, 919, 920]])
# clean_dataset
clean_dataset = combined_study.loc[combined_study["Mouse ID"] != "g989"]
clean_dataset
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

regimen_mean = clean_dataset.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
regimen_median = clean_dataset.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
regimen_var = clean_dataset.groupby('Drug Regimen').var()['Tumor Volume (mm3)']  
regimen_std = clean_dataset.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
regimen_sem = clean_dataset.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

# Format decimal places
pd.options.display.float_format = '{:,.2f}'.format

statistics_summary_df = pd.DataFrame({'Mean':regimen_mean, 'Median':regimen_median, 'Variance':regimen_var, 
                                      'Standard Deviation':regimen_std, 'Standard Error of Mean':regimen_sem})
statistics_summary_df

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
single_group_by = clean_dataset.groupby('Drug Regimen')

# Format decimal places
pd.options.display.float_format = '{:,.2f}'.format

# Use aggregate method to produce summary statistics
summary_stats_table_2 = single_group_by.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
summary_stats_table_2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_dataset.groupby("Mouse ID").max()["Timepoint"] 
max_df = max_timepoint.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merge_df = pd.merge(max_df, clean_dataset, on=["Mouse ID", "Timepoint"], how="left")
merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for x in drug_list:
   
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    drug_volume = merge_df.loc[merge_df["Drug Regimen"] == x]["Tumor Volume (mm3)"]
    
    # add subset 
   
    tumor_volume.append(drug_volume)   

    # Determine outliers using upper and lower bounds
    
    quartiles = drug_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    drug_outlier = drug_volume.loc[(drug_volume < lower_bound) | (drug_volume > upper_bound)]
    print(x, drug_outlier)
    

    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_volume, labels=drug_list)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
fig1, ax1 = plt.subplots(figsize=(10, 5))
marker_size=10
plt.scatter(avg_vol_wt['Tumor Volume (mm3)'], avg_vol_wt['Weight (g)'], s=50, color="purple")
plt.title('Average Tumor Volume Versus Weight',fontsize =20)
plt.xlabel('Average Tumor Volume (mm3)',fontsize =14)
plt.ylabel('Weight (g)',fontsize =14)

plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr = round(st.pearsonr(avg_vol_wt['Weight (g)'],avg_vol_wt['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume for Caplomulin is {corr}")

```

```python
# Finding linear regression line and plotting it

# Creating x-axis and y-axis values

x_values = avg_vol_wt['Tumor Volume (mm3)']
y_values = avg_vol_wt['Weight (g)']

# Plotting scatter plot and regression line
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values, color="purple")
plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq,(41
                      ,17),fontsize=15,color="red")
plt.title('Average Tumor Volume Versus Weight',fontsize =20)
plt.xlabel('Avg Tumor Volume',fontsize =14)
plt.ylabel('Weight (g)',fontsize =14)
plt.savefig("matplotlib")

# print(f"slope:{slope}")
# print(f"intercept:{intercept}")
# print(f"rvalue (Correlation coefficient):{rvalue}")
# print(f"r-squared: {rvalue**2}")
# print(f"stderr:{stderr}")

```
